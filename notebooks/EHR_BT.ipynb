## Round 01 - XGBoost Basic

````python
# =========================================================
# üöÄ XGBOOST TRAINING PIPELINE (com prefixo por round)
# =========================================================
# Prefixo: usado em todos os arquivos salvos no Google Drive
# Exemplo: r1_xgb_best_model_..., r1_classification_report_...
# =========================================================

import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    precision_score, recall_score, f1_score, roc_auc_score, log_loss
)
from xgboost import XGBClassifier
import csv

# ---------------------------------------------------------
# 1Ô∏è‚É£ CONFIGURA√á√ïES GERAIS
# ---------------------------------------------------------
drive_path = "/content/drive/MyDrive/AI-HealthCare/EHR/heart-disease-fairness"

# Prefixo e identifica√ß√£o do round
prefixo = "r1"  # üîπ prefixo padr√£o (ex: "r1", "r2", "r3")
round_id = "round_01"

# Caminhos principais
data_dir = f"{drive_path}/data/splits"
models_base = f"{drive_path}/models/xgboost/rounds"
results_dir = f"{drive_path}/results"
round_path = os.path.join(models_base, round_id)

# Cria√ß√£o de diret√≥rios
os.makedirs(data_dir, exist_ok=True)
os.makedirs(models_base, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)
os.makedirs(round_path, exist_ok=True)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
print(f"üöÄ Iniciando treino XGBoost | {round_id} | {timestamp}")

# ---------------------------------------------------------
# 2Ô∏è‚É£ CARREGAR DADOS
# ---------------------------------------------------------
train_path = os.path.join(data_dir, "train.csv")
test_path = os.path.join(data_dir, "test.csv")

print(f"üìÇ Lendo datasets:\n ‚Üí Train: {train_path}\n ‚Üí Test:  {test_path}")
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

target_col = "num"
df_train["label"] = (df_train[target_col] > 0).astype(int)
df_test["label"] = (df_test[target_col] > 0).astype(int)

X_train = df_train.drop(columns=[target_col, "label"])
y_train = df_train["label"]
X_test = df_test.drop(columns=[target_col, "label"])
y_test = df_test["label"]

print(f"‚úÖ Train shape: {X_train.shape} | Test shape: {X_test.shape}")

# ---------------------------------------------------------
# 3Ô∏è‚É£ TREINAMENTO DO MODELO
# ---------------------------------------------------------
xgb_model = XGBClassifier(
    objective="binary:logistic",
    tree_method="hist",
    device="cuda",
    eval_metric=["logloss", "error"],
    n_estimators=400,
    learning_rate=0.03,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=1.2,
    random_state=42,
    early_stopping_rounds=30
)

print("\nüèãÔ∏è‚Äç‚ôÇÔ∏è Iniciando treino...")
eval_set = [(X_train, y_train), (X_test, y_test)]
xgb_model.fit(X_train, y_train, eval_set=eval_set, verbose=50)

best_iteration = getattr(xgb_model, "best_iteration", None)
print(f"üèÜ Melhor itera√ß√£o: {best_iteration if best_iteration else 'N/A'}")

# ---------------------------------------------------------
# 4Ô∏è‚É£ AVALIA√á√ÉO DO MODELO
# ---------------------------------------------------------
y_pred = xgb_model.predict(X_test)
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]

metrics = {
    "accuracy": round(accuracy_score(y_test, y_pred), 4),
    "precision": round(precision_score(y_test, y_pred, zero_division=0), 4),
    "recall": round(recall_score(y_test, y_pred, zero_division=0), 4),
    "f1_score": round(f1_score(y_test, y_pred, zero_division=0), 4),
    "auc_roc": round(roc_auc_score(y_test, y_pred_proba), 4),
    "log_loss": round(log_loss(y_test, y_pred_proba), 4),
    "train_accuracy": round(accuracy_score(y_train, xgb_model.predict(X_train)), 4)
}

# ---------------------------------------------------------
# 5Ô∏è‚É£ SALVAMENTO DOS ARQUIVOS (com prefixo)
# ---------------------------------------------------------
model_path = os.path.join(round_path, f"{prefixo}_xgb_best_model_{timestamp}.json")
xgb_model.save_model(model_path)
print(f"üíæ Modelo salvo: {model_path}")

# Classification Report
report_path = os.path.join(round_path, f"{prefixo}_classification_report_{timestamp}.txt")
report = classification_report(y_test, y_pred, zero_division=0)
cm = confusion_matrix(y_test, y_pred)
with open(report_path, "w") as f:
    f.write("=== CLASSIFICATION REPORT ===\n")
    f.write(report)
    f.write("\n=== CONFUSION MATRIX ===\n")
    f.write(str(cm))
print(f"üìÑ {prefixo}_classification_report.txt salvo")

# Training Log
results = xgb_model.evals_result()
epochs = len(results["validation_0"]["logloss"])
log_path = os.path.join(round_path, f"{prefixo}_train_log_{timestamp}.txt")
with open(log_path, "w") as f:
    f.write("epoch\ttrain_logloss\tval_logloss\ttrain_error\tval_error\n")
    for i in range(epochs):
        f.write(f"{i}\t{results['validation_0']['logloss'][i]:.4f}\t"
                f"{results['validation_1']['logloss'][i]:.4f}\t"
                f"{results['validation_0']['error'][i]:.4f}\t"
                f"{results['validation_1']['error'][i]:.4f}\n")
print(f"üìÑ {prefixo}_train_log.txt salvo")

# Round Summary
summary_path = os.path.join(round_path, f"{prefixo}_round_summary_{timestamp}.txt")
with open(summary_path, "w") as f:
    f.write(f"==============================\n")
    f.write(f"XGBOOST ROUND SUMMARY\n")
    f.write(f"==============================\n")
    f.write(f"Round ID: {round_id}\n")
    f.write(f"Prefixo: {prefixo}\n")
    f.write(f"Timestamp: {timestamp}\n\n")
    f.write(f"üîß Modelo: XGBoostClassifier\n")
    f.write(f"Treino: {len(X_train)} amostras\n")
    f.write(f"Valida√ß√£o: {len(X_test)} amostras\n\n")
    f.write("üìä M√âTRICAS\n")
    for k, v in metrics.items():
        f.write(f"{k}: {v}\n")
    f.write(f"\nüèãÔ∏è‚Äç‚ôÇÔ∏è Best Iteration: {best_iteration}\n")
    f.write(f"üíæ Modelo: {model_path}\n")
print(f"üìÑ {prefixo}_round_summary.txt salvo")

# ---------------------------------------------------------
# 6Ô∏è‚É£ CONSOLIDAR EM results.txt
# ---------------------------------------------------------
fields = [
    "timestamp", "prefix", "round", "model",
    "accuracy", "precision", "recall", "f1_score", "auc_roc", "log_loss",
    "train_accuracy", "train_samples", "test_samples",
    "best_iteration", "model_path"
]

row_data = {
    "timestamp": timestamp,
    "prefix": prefixo,
    "round": round_id,
    "model": "xgboost",
    "accuracy": metrics["accuracy"],
    "precision": metrics["precision"],
    "recall": metrics["recall"],
    "f1_score": metrics["f1_score"],
    "auc_roc": metrics["auc_roc"],
    "log_loss": metrics["log_loss"],
    "train_accuracy": metrics["train_accuracy"],
    "train_samples": len(X_train),
    "test_samples": len(X_test),
    "best_iteration": best_iteration if best_iteration is not None else "",
    "model_path": model_path,
}

results_txt = os.path.join(results_dir, "results.txt")
write_header = not os.path.exists(results_txt)

with open(results_txt, "a", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields, delimiter="\t")
    if write_header:
        writer.writeheader()
    writer.writerow(row_data)

print(f"üìò results.txt atualizado em: {results_txt}")

# ---------------------------------------------------------
# 7Ô∏è‚É£ FINALIZA√á√ÉO
# ---------------------------------------------------------
print(f"""
üèÅ Round {round_id} conclu√≠do!
---------------------------------------
üìä Acur√°cia: {metrics['accuracy']}
üè∑Ô∏è Prefixo: {prefixo}
üìÅ Sa√≠da: {round_path}
üìÑ Consolidado: {results_txt}
""")
````

## Round_02 - Ajustes de parametros apenas

| Par√¢metro          | round_01 | round_02 proposto | Motivo                           |
|--------------------|----------|-------------------|----------------------------------|
| scale_pos_weight   | 1.2      | 1.8               | aumenta sensibilidade para reduzir FN |
| max_depth          | 4        | 5                 | melhora poder de intera√ß√£o       |
| Threshold          | 0.50     | 0.55              | reduz falsos positivos           |


üí° Tudo o resto igual: learning_rate=0.03, subsample=0.8, colsample_bytree=0.8.

````python
# =========================================================
# üöÄ XGBOOST TRAINING PIPELINE ‚Äî ROUND CONTROLADO (com prefixo r2)
# =========================================================
# Altera√ß√µes:
#  ‚Ä¢ scale_pos_weight = 1.8
#  ‚Ä¢ max_depth = 5
#  ‚Ä¢ threshold = 0.55
#  ‚Ä¢ arquivos gravados com prefixo "r2_"
# =========================================================

import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    precision_score, recall_score, f1_score, roc_auc_score, log_loss
)
from xgboost import XGBClassifier
import csv

# ---------------------------------------------------------
# 1Ô∏è‚É£ CONFIGURA√á√ïES GERAIS
# ---------------------------------------------------------
drive_path = "/content/drive/MyDrive/AI-HealthCare/EHR/heart-disease-fairness"

# Prefixo do round
prefixo = "r2"  # üîπ todos os arquivos salvos ter√£o esse prefixo

# Caminhos principais
data_dir = f"{drive_path}/data/splits"
models_base = f"{drive_path}/models/xgboost/rounds"
results_dir = f"{drive_path}/results"

# Cria√ß√£o de diret√≥rios
os.makedirs(data_dir, exist_ok=True)
os.makedirs(models_base, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

# Identifica√ß√£o do round
round_id = "round_02"
round_path = os.path.join(models_base, round_id)
os.makedirs(round_path, exist_ok=True)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
print(f"üöÄ Iniciando treino XGBoost | {round_id} | {timestamp}")

# ---------------------------------------------------------
# 2Ô∏è‚É£ CARREGAR DADOS
# ---------------------------------------------------------
train_path = os.path.join(data_dir, "train.csv")
test_path = os.path.join(data_dir, "test.csv")

print(f"üìÇ Lendo datasets:\n ‚Üí Train: {train_path}\n ‚Üí Test:  {test_path}")
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# Criar label bin√°rio
target_col = "num"
df_train["label"] = (df_train[target_col] > 0).astype(int)
df_test["label"] = (df_test[target_col] > 0).astype(int)

X_train = df_train.drop(columns=[target_col, "label"])
y_train = df_train["label"]
X_test = df_test.drop(columns=[target_col, "label"])
y_test = df_test["label"]

print(f"‚úÖ Train shape: {X_train.shape} | Test shape: {X_test.shape}")

# ---------------------------------------------------------
# 3Ô∏è‚É£ CONFIGURAR E TREINAR MODELO XGBOOST
# ---------------------------------------------------------
xgb_model = XGBClassifier(
    objective="binary:logistic",
    tree_method="hist",
    device="cuda",
    eval_metric=["logloss", "error"],
    n_estimators=400,
    learning_rate=0.03,
    max_depth=5,  # üîπ aumentado
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=1.8,  # üîπ aumentado
    random_state=42,
    early_stopping_rounds=30
)

print("\nüèãÔ∏è‚Äç‚ôÇÔ∏è Iniciando treino...")
eval_set = [(X_train, y_train), (X_test, y_test)]
xgb_model.fit(X_train, y_train, eval_set=eval_set, verbose=50)

best_iteration = getattr(xgb_model, "best_iteration", None)
print(f"üèÜ Melhor itera√ß√£o: {best_iteration if best_iteration else 'N/A'}")

# ---------------------------------------------------------
# 4Ô∏è‚É£ AVALIA√á√ÉO COM THRESHOLD AJUSTADO
# ---------------------------------------------------------
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]
threshold = 0.55  # üîπ novo limite de decis√£o
y_pred = (y_pred_proba >= threshold).astype(int)

metrics = {
    "accuracy": round(accuracy_score(y_test, y_pred), 4),
    "precision": round(precision_score(y_test, y_pred, zero_division=0), 4),
    "recall": round(recall_score(y_test, y_pred, zero_division=0), 4),
    "f1_score": round(f1_score(y_test, y_pred, zero_division=0), 4),
    "auc_roc": round(roc_auc_score(y_test, y_pred_proba), 4),
    "log_loss": round(log_loss(y_test, y_pred_proba), 4),
    "train_accuracy": round(accuracy_score(y_train, xgb_model.predict(X_train)), 4)
}

# ---------------------------------------------------------
# 5Ô∏è‚É£ SALVAR MODELO E RELAT√ìRIOS (com prefixo)
# ---------------------------------------------------------
model_path = os.path.join(round_path, f"{prefixo}_xgb_best_model_{timestamp}.json")
xgb_model.save_model(model_path)
print(f"üíæ Modelo salvo: {model_path}")

# Classification report
report_path = os.path.join(round_path, f"{prefixo}_classification_report_{timestamp}.txt")
report = classification_report(y_test, y_pred, zero_division=0)
cm = confusion_matrix(y_test, y_pred)
with open(report_path, "w") as f:
    f.write("=== CLASSIFICATION REPORT ===\n")
    f.write(report)
    f.write("\n=== CONFUSION MATRIX ===\n")
    f.write(str(cm))
print(f"üìÑ Report salvo: {report_path}")

# Training log
log_path = os.path.join(round_path, f"{prefixo}_train_log_{timestamp}.txt")
results = xgb_model.evals_result()
epochs = len(results["validation_0"]["logloss"])
with open(log_path, "w") as f:
    f.write("epoch\ttrain_logloss\tval_logloss\ttrain_error\tval_error\n")
    for i in range(epochs):
        f.write(f"{i}\t{results['validation_0']['logloss'][i]:.4f}\t"
                f"{results['validation_1']['logloss'][i]:.4f}\t"
                f"{results['validation_0']['error'][i]:.4f}\t"
                f"{results['validation_1']['error'][i]:.4f}\n")
print(f"üìÑ Log salvo: {log_path}")

# ---------------------------------------------------------
# 6Ô∏è‚É£ SUM√ÅRIO DO ROUND
# ---------------------------------------------------------
summary_path = os.path.join(round_path, f"{prefixo}_round_summary_{timestamp}.txt")
with open(summary_path, "w") as f:
    f.write(f"==============================\n")
    f.write(f"XGBOOST ROUND SUMMARY\n")
    f.write(f"==============================\n")
    f.write(f"Round ID: {round_id}\n")
    f.write(f"Prefixo: {prefixo}\n")
    f.write(f"Timestamp: {timestamp}\n\n")
    f.write(f"üîß Modelo: XGBoostClassifier\n")
    f.write(f"Treino: {len(X_train)} amostras\n")
    f.write(f"Valida√ß√£o: {len(X_test)} amostras\n\n")
    f.write("üìä M√âTRICAS\n")
    for k, v in metrics.items():
        f.write(f"{k}: {v}\n")
    f.write(f"\nüèãÔ∏è‚Äç‚ôÇÔ∏è Best Iteration: {best_iteration}\n")
    f.write(f"Threshold: {threshold}\n")
    f.write(f"üíæ Modelo: {model_path}\n")
print(f"üìÑ Sum√°rio salvo: {summary_path}")

# ---------------------------------------------------------
# 7Ô∏è‚É£ ATUALIZAR results.txt CONSOLIDADO
# ---------------------------------------------------------
results_txt = os.path.join(results_dir, "results.txt")
fields = [
    "timestamp", "prefix", "round", "model",
    "accuracy", "precision", "recall", "f1_score", "auc_roc", "log_loss",
    "train_accuracy", "train_samples", "test_samples",
    "best_iteration", "threshold", "model_path"
]
row_data = {
    "timestamp": timestamp,
    "prefix": prefixo,
    "round": round_id,
    "model": "xgboost",
    "accuracy": metrics["accuracy"],
    "precision": metrics["precision"],
    "recall": metrics["recall"],
    "f1_score": metrics["f1_score"],
    "auc_roc": metrics["auc_roc"],
    "log_loss": metrics["log_loss"],
    "train_accuracy": metrics["train_accuracy"],
    "train_samples": len(X_train),
    "test_samples": len(X_test),
    "best_iteration": best_iteration if best_iteration is not None else "",
    "threshold": threshold,
    "model_path": model_path,
}
write_header = not os.path.exists(results_txt)
with open(results_txt, "a", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields, delimiter="\t")
    if write_header:
        writer.writeheader()
    writer.writerow(row_data)
print(f"üìò results.txt atualizado em: {results_txt}")

# ---------------------------------------------------------
# 8Ô∏è‚É£ FINALIZA√á√ÉO
# ---------------------------------------------------------
print(f"""
üèÅ Round {round_id} conclu√≠do!
---------------------------------------
üìä Acur√°cia: {metrics['accuracy']}
üè∑Ô∏è Prefixo: {prefixo}
üìÅ Sa√≠da: {round_path}
üìÑ Consolidado: {results_txt}
""")
````

````python
# =========================================================
# üöÄ XGBOOST TRAINING PIPELINE ‚Äî ROUND_03 (com prefixo r3)
# =========================================================
# Altera√ß√µes:
#  ‚Ä¢ aumento na quantidade de registros do treino
#  ‚Ä¢ oversampling direcionado em FP e FN
# =========================================================

import pandas as pd
import numpy as np
import os
from datetime import datetime
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix,
    precision_score, recall_score, f1_score, roc_auc_score, log_loss
)
from xgboost import XGBClassifier
import csv

# ---------------------------------------------------------
# 1Ô∏è‚É£ CONFIGURA√á√ïES GERAIS
# ---------------------------------------------------------
drive_path = "/content/drive/MyDrive/AI-HealthCare/EHR/heart-disease-fairness"

# Prefixo do round
prefixo = "r3"  # üîπ prefixo de todos os arquivos

# Caminhos principais
data_dir = f"{drive_path}/data/splits"
models_base = f"{drive_path}/models/xgboost/rounds"
results_dir = f"{drive_path}/results"

# Cria√ß√£o de diret√≥rios
os.makedirs(data_dir, exist_ok=True)
os.makedirs(models_base, exist_ok=True)
os.makedirs(results_dir, exist_ok=True)

# Identifica√ß√£o do round
round_id = "round_03"
round_path = os.path.join(models_base, round_id)
os.makedirs(round_path, exist_ok=True)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
print(f"üöÄ Iniciando treino XGBoost | {round_id} | {timestamp}")

# ---------------------------------------------------------
# 2Ô∏è‚É£ CARREGAR DADOS
# ---------------------------------------------------------
train_path = os.path.join(data_dir, "round_02_train.csv")
test_path = os.path.join(data_dir, "round_02_test.csv")

print(f"üìÇ Lendo datasets:\n ‚Üí Train: {train_path}\n ‚Üí Test:  {test_path}")
df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# Criar label bin√°rio
target_col = "num"
df_train["label"] = (df_train[target_col] > 0).astype(int)
df_test["label"] = (df_test[target_col] > 0).astype(int)

X_train = df_train.drop(columns=[target_col, "label"])
y_train = df_train["label"]
X_test = df_test.drop(columns=[target_col, "label"])
y_test = df_test["label"]

print(f"‚úÖ Train shape: {X_train.shape} | Test shape: {X_test.shape}")

# ---------------------------------------------------------
# 3Ô∏è‚É£ CONFIGURAR E TREINAR MODELO XGBOOST
# ---------------------------------------------------------
xgb_model = XGBClassifier(
    objective="binary:logistic",
    tree_method="hist",
    device="cuda",
    eval_metric=["logloss", "error"],
    n_estimators=400,
    learning_rate=0.03,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=1.8,
    random_state=42,
    early_stopping_rounds=30
)

print("\nüèãÔ∏è‚Äç‚ôÇÔ∏è Iniciando treino...")
eval_set = [(X_train, y_train), (X_test, y_test)]
xgb_model.fit(X_train, y_train, eval_set=eval_set, verbose=50)

best_iteration = getattr(xgb_model, "best_iteration", None)
print(f"üèÜ Melhor itera√ß√£o: {best_iteration if best_iteration else 'N/A'}")

# ---------------------------------------------------------
# 4Ô∏è‚É£ AVALIA√á√ÉO COM THRESHOLD AJUSTADO
# ---------------------------------------------------------
y_pred_proba = xgb_model.predict_proba(X_test)[:, 1]
threshold = 0.55  # üîπ novo limite de decis√£o
y_pred = (y_pred_proba >= threshold).astype(int)

metrics = {
    "accuracy": round(accuracy_score(y_test, y_pred), 4),
    "precision": round(precision_score(y_test, y_pred, zero_division=0), 4),
    "recall": round(recall_score(y_test, y_pred, zero_division=0), 4),
    "f1_score": round(f1_score(y_test, y_pred, zero_division=0), 4),
    "auc_roc": round(roc_auc_score(y_test, y_pred_proba), 4),
    "log_loss": round(log_loss(y_test, y_pred_proba), 4),
    "train_accuracy": round(accuracy_score(y_train, xgb_model.predict(X_train)), 4)
}

# ---------------------------------------------------------
# 5Ô∏è‚É£ SALVAR MODELO E RELAT√ìRIOS (com prefixo r3_)
# ---------------------------------------------------------
model_path = os.path.join(round_path, f"{prefixo}_xgb_best_model_{timestamp}.json")
xgb_model.save_model(model_path)
print(f"üíæ Modelo salvo: {model_path}")

# Classification report
report_path = os.path.join(round_path, f"{prefixo}_classification_report_{timestamp}.txt")
report = classification_report(y_test, y_pred, zero_division=0)
cm = confusion_matrix(y_test, y_pred)
with open(report_path, "w") as f:
    f.write("=== CLASSIFICATION REPORT ===\n")
    f.write(report)
    f.write("\n=== CONFUSION MATRIX ===\n")
    f.write(str(cm))
print(f"üìÑ Report salvo: {report_path}")

# Training log
log_path = os.path.join(round_path, f"{prefixo}_train_log_{timestamp}.txt")
results = xgb_model.evals_result()
epochs = len(results["validation_0"]["logloss"])
with open(log_path, "w") as f:
    f.write("epoch\ttrain_logloss\tval_logloss\ttrain_error\tval_error\n")
    for i in range(epochs):
        f.write(f"{i}\t{results['validation_0']['logloss'][i]:.4f}\t"
                f"{results['validation_1']['logloss'][i]:.4f}\t"
                f"{results['validation_0']['error'][i]:.4f}\t"
                f"{results['validation_1']['error'][i]:.4f}\n")
print(f"üìÑ Log salvo: {log_path}")

# ---------------------------------------------------------
# 6Ô∏è‚É£ SUM√ÅRIO DO ROUND
# ---------------------------------------------------------
summary_path = os.path.join(round_path, f"{prefixo}_round_summary_{timestamp}.txt")
with open(summary_path, "w") as f:
    f.write(f"==============================\n")
    f.write(f"XGBOOST ROUND SUMMARY\n")
    f.write(f"==============================\n")
    f.write(f"Round ID: {round_id}\n")
    f.write(f"Prefixo: {prefixo}\n")
    f.write(f"Timestamp: {timestamp}\n\n")
    f.write(f"üîß Modelo: XGBoostClassifier\n")
    f.write(f"Treino: {len(X_train)} amostras\n")
    f.write(f"Valida√ß√£o: {len(X_test)} amostras\n\n")
    f.write("üìä M√âTRICAS\n")
    for k, v in metrics.items():
        f.write(f"{k}: {v}\n")
    f.write(f"\nüèãÔ∏è‚Äç‚ôÇÔ∏è Best Iteration: {best_iteration}\n")
    f.write(f"Threshold: {threshold}\n")
    f.write(f"üíæ Modelo: {model_path}\n")
print(f"üìÑ Sum√°rio salvo: {summary_path}")

# ---------------------------------------------------------
# 7Ô∏è‚É£ ATUALIZAR results.txt CONSOLIDADO
# ---------------------------------------------------------
results_txt = os.path.join(results_dir, "results.txt")
fields = [
    "timestamp", "prefix", "round", "model",
    "accuracy", "precision", "recall", "f1_score", "auc_roc", "log_loss",
    "train_accuracy", "train_samples", "test_samples",
    "best_iteration", "threshold", "model_path"
]
row_data = {
    "timestamp": timestamp,
    "prefix": prefixo,
    "round": round_id,
    "model": "xgboost",
    "accuracy": metrics["accuracy"],
    "precision": metrics["precision"],
    "recall": metrics["recall"],
    "f1_score": metrics["f1_score"],
    "auc_roc": metrics["auc_roc"],
    "log_loss": metrics["log_loss"],
    "train_accuracy": metrics["train_accuracy"],
    "train_samples": len(X_train),
    "test_samples": len(X_test),
    "best_iteration": best_iteration if best_iteration is not None else "",
    "threshold": threshold,
    "model_path": model_path,
}
write_header = not os.path.exists(results_txt)
with open(results_txt, "a", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=fields, delimiter="\t")
    if write_header:
        writer.writeheader()
    writer.writerow(row_data)
print(f"üìò results.txt atualizado em: {results_txt}")

# ---------------------------------------------------------
# 8Ô∏è‚É£ FINALIZA√á√ÉO
# ---------------------------------------------------------
print(f"""
üèÅ Round {round_id} conclu√≠do!
---------------------------------------
üìä Acur√°cia: {metrics['accuracy']}
üè∑Ô∏è Prefixo: {prefixo}
üìÅ Sa√≠da: {round_path}
üìÑ Consolidado: {results_txt}
""")
````
